```python
# importar librerías necesarias para la realización de la asignación

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# leer los datos del xlsx

archivo = 'Datos_primer_TP_19Co2024_i0907.xlsx'
xlsx = pd.read_excel(archivo)
```

```python
# cambiar de formato de fecha la columna "Fecha"

xlsx['Fecha'] = pd.to_datetime(xlsx['Fecha'], format='%d/%m/%Y')
```

```python
# extrar el año de la columna "Fecha"

xlsx['Año'] = xlsx['Fecha'].dt.year


# filtrar los datos por año

ventas_2021 = xlsx[xlsx['Año'] == 2021]['Ventas']
ventas_2022 = xlsx[xlsx['Año'] == 2022]['Ventas']
ventas_2023 = xlsx[xlsx['Año'] == 2023]['Ventas']
```

```python
# definir la funcion que grafica la distribución acumulativa empírica

def plot_ecdf(data, year):
    plt.figure(figsize=(8, 5))
    sns.ecdfplot(data, label=f'Año {year}')
    plt.title(f'Función Empírica de Distribución (ECDF) - Año {year}')
    plt.xlabel('Ventas ($)')
    plt.ylabel('Probabilidad acumulada')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
plot_ecdf(ventas_2021, 2021)
```

```python
plot_ecdf(ventas_2022, 2022)
```

```python
plot_ecdf(ventas_2023, 2023)
```

```python
# definir la función que grafica distribuciones utilizando la estimación de densidad

def plot_kde(data, year):
    plt.figure(figsize=(8, 5))
    sns.histplot(data, bins=30, stat='density', label='Histograma') 
    sns.kdeplot(data, fill=True, label=f'Año {year}') 
    plt.title(f'Aproximación de la Densidad de Ventas - Año {year}')
    plt.xlabel('Ventas ($)')
    plt.ylabel('Densidad')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
plot_kde(ventas_2021, 2021)
```

```python
plot_kde(ventas_2022, 2022)
```

```python
plot_kde(ventas_2023, 2023)
```
